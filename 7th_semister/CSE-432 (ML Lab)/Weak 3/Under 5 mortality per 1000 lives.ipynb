```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style

year_x_axis = [1993, 1997, 2001, 2004, 2007, 2011, 2014, 2017, 2019, 2019, 2022, 2030]
death_y_axis = [133, 116, 94, 86, 65, 53, 46, 45, 31, 40, 34, 25]

plt.pie(year_x_axis, labels=year_x_axis)
plt.show()
#x_pos = [0,0.3,2,4.5,5.5]
style.use("seaborn-v0_8-poster")
plt.bar(year_x_axis, death_y_axis, width = 2.1)
plt.title('Unser 5 mortality rate (U5MR) per 1,000 live births')
plt.xlabel('Year')
plt.ylabel('U5MR')

plt.xticks(rotation=45)
plt.show()
```
